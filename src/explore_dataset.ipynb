```python
import os
import matplotlib.pyplot as plt

train_dir = "../Data/train"  
cats_dir = os.path.join(train_dir, "cats")
dogs_dir = os.path.join(train_dir, "dogs")

num_cats = len([f for f in os.listdir(cats_dir) if os.path.isfile(os.path.join(cats_dir, f))])
num_dogs = len([f for f in os.listdir(dogs_dir) if os.path.isfile(os.path.join(dogs_dir, f))])

print("Number of training images:")
print(f"Cats: {num_cats}")
print(f"Dogs: {num_dogs}")


# bar chart
classes = ["Cats", "Dogs"]
counts = [num_cats, num_dogs]

plt.figure(figsize=(5, 4))
plt.bar(classes, counts, color=["green", "blue"])
plt.title("Class distribution in TRAIN set")
plt.xlabel("Class")
plt.ylabel("Number of images")
for i, c in enumerate(counts):
    plt.text(i, c + 10, str(c), ha="center")
plt.tight_layout()
plt.show()
```
